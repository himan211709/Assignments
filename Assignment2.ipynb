```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv(f"D:\Data Science\House Price India.csv")
```

univariate<br>
histogram of House Prices

```python
plt.hist(data['Price'], bins=20)
plt.xlabel('House Price')
plt.ylabel('Frequency')
plt.title('Histogram of House Prices')
plt.show()
#Bar plot of Number of Bedrooms
sns.countplot(x='number of bedrooms', data=data)
plt.xlabel('Number of Bedrooms')
plt.ylabel('Count')
plt.title('Bar Plot of Number of Bedrooms')
plt.show()
```

bivariate<br>
scatter plot of Living Area vs. Price

```python
plt.scatter(data['living area'], data['Price'])
plt.xlabel('Living Area')
plt.ylabel('House Price')
plt.title('Scatter Plot of Living Area vs. Price')
plt.show()
#Box plot of Number of Bedrooms vs. Price
sns.boxplot(x='number of bedrooms', y='Price', data=data)
plt.xlabel('Number of Bedrooms')
plt.ylabel('House Price')
plt.title('Box Plot of Number of Bedrooms vs. Price')
plt.show()
```

multivariate<br>
heatmap

```python
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
sns.pairplot(data[['living area', 'lot area', 'number of bedrooms', 'number of bathrooms', 'Price']])
plt.title('Pair Plot of Features')
plt.show()
```

descriptive stats

```python
descriptive_stats = data.describe()
print(descriptive_stats)
#No null values
missing_values = data.isnull().sum()
print(missing_values)
```
